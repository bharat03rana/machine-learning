Lung disease prediction

```python
#import libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
#loading dataset
lung_disease_data=pd.read_csv('/content/dataset_med.csv')
```

```python
#first 5 rows
lung_disease_data.head()
```

```python
#tail
lung_disease_data.tail()
```

```python
#shape
lung_disease_data.shape
```

```python
#info about data
lung_disease_data.info()
```

```python
#checking for missing values
lung_disease_data.isnull().sum()
```

```python
#checking distribution of servival value
lung_disease_data.survived.value_counts()
```

0--> not survived
1--> survived

```python
#label encoder
le=LabelEncoder()
lung_disease_data['gender']=le.fit_transform(lung_disease_data['gender'])
```

```python
lung_disease_data['country']=le.fit_transform(lung_disease_data['country'])
```

```python
lung_disease_data['cancer_stage']=le.fit_transform(lung_disease_data['cancer_stage'])
```

```python
lung_disease_data['family_history']=le.fit_transform(lung_disease_data['family_history'])
```

```python
lung_disease_data['smoking_status']=le.fit_transform(lung_disease_data['smoking_status'])
```

```python
lung_disease_data['treatment_type']=le.fit_transform(lung_disease_data['treatment_type'])
```

```python
lung_disease_data['diagnosis_date']=le.fit_transform(lung_disease_data['diagnosis_date'])
```

```python
lung_disease_data['end_treatment_date']=le.fit_transform(lung_disease_data['end_treatment_date'])
```

```python
lung_disease_data.head()
```

```python
#splitting features and target
X=lung_disease_data.drop(columns=['survived','id'],axis=1)
Y=lung_disease_data['survived']
```

```python
print(X)
```

```python
print(Y)
```

```python
#splitting data into training
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

model training

```python
#Logistic regression
model=LogisticRegression()
model=model.fit(X_train,Y_train)
```

model evaluation

```python
#accuracy score on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(Y_train,X_train_prediction)
print('Accuracy score of training data:',training_data_accuracy)
```

```python
#accuracy score on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(Y_test,X_test_prediction)
print('Accuracy score of test data:',test_data_accuracy)
```

Building a predictive model

```python
input_data=(37.0,1,17,3467,0,0,3,19.7,178,0,0,0,0,1,3688)
#change the input data to numpy array
input_data_as_numpy_array=np.asarray(input_data)

#reshape the numpy array as we are predicting for one instance
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)
```

```python
prediction=model.predict(input_data_reshaped)
print(prediction)
```

```python
if prediction[0]==0:
  print('not survived')
else:
  print('survived')
```
